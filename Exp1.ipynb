```python
# Implement and Demonstrate Find-S Algorithm for Finding the most Specific Hypothesis.

import csv

# Print the most general and most specific hypotheses
print("\nThe most general hypothesis: ['?','?','?','?','?','?']\n")
print("\nThe most specific hypothesis: ['0','0','0','0','0','0']\n")

# Initialize an empty list to store the training data
a = []

# Read the training data from the CSV file
print("\nThe given Training Data Set:")
with open('C:/Users/Anush Bhandary/Downloads/enjoysport.csv', 'r') as csvFile:
    reader = csv.reader(csvFile)
    for row in reader:
        a.append(row)
        print(row)

# Determine the number of attributes in the dataset
num_attributes = len(a[0]) - 1

# Print the initial value of the hypothesis
hypothesis = ['0'] * num_attributes
print('\nInitial Value of the hypothesis: ')
print(hypothesis)

# Initialize the hypothesis with the first positive example
for j in range(num_attributes):
    hypothesis[j] = a[1][j]

# Print the process of finding the maximally specific hypothesis
print('\nFind S: Finding Maximally specific hypothesis\n')
for i in range(1, len(a)):
    if a[i][num_attributes] == 'yes':
        for j in range(num_attributes):
            if a[i][j] != hypothesis[j]:
                hypothesis[j] = '?'
            else:
                hypothesis[j] = a[i][j]
    print(f"For training example no: {i} the hypothesis is {hypothesis}")

# Print the maximally specific hypothesis
print("\nThe Maximally Specific Hypothesis for a given Training Examples:\n")
print(hypothesis)
```
