```python
# Implement and Demonstrate the working of Decision Tree Algorithm

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree

# Load the iris dataset
iris = load_iris()

# Create a DataFrame from the iris data
data = pd.DataFrame(data=iris.data, columns=iris.feature_names)

# Add 'species' column to the DataFrame
data['species'] = iris.target

# Map numerical species to their respective names
target_names = iris.target_names
target_dict = dict(zip(range(len(target_names)), target_names))
data['species'] = data['species'].map(target_dict)

# Separate features and target variable
X = data.drop(columns="species")
y = data["species"]

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=93)

# Initialize and fit the Decision Tree Classifier
dtc = DecisionTreeClassifier(max_depth=3, random_state=93)
dtc.fit(X_train, y_train)

# Plot the Decision Tree
plt.figure(figsize=(15, 10))
plot_tree(dtc, feature_names=iris.feature_names, class_names=target_names, rounded=True, filled=True, fontsize=12)
plt.show()

# Make predictions on the test set
y_pred = dtc.predict(X_test)

# Print accuracy on the test set
accuracy = np.mean(y_pred == y_test)
print("Accuracy:", accuracy)
```
