```python
# Demonstrate the working of SVM classifier for a Suitable Dataset

import numpy as np 
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

# Load the Iris dataset
iris = datasets.load_iris()

# Select the first two features for visualization
X = iris.data[:, :2]
y = iris.target

# Create and train the SVM model with a linear kernel
svm = SVC(kernel='linear', C=1)
svm.fit(X, y)

# Scatter plot of the Iris dataset
plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.coolwarm)
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Scatter plot of Iris dataset')
plt.colorbar(label='Target')

# Decision boundary
w = svm.coef_[0]
a = -w[0] / w[1]
xx = np.linspace(X[:, 0].min(), X[:, 0].max())
yy = a * xx - svm.intercept_[0] / w[1]

plt.plot(xx, yy, 'k-')  # Decision boundary
yy_down = yy - 1 / w[1]
yy_up = yy + 1 / w[1]
plt.plot(xx, yy_down, 'k--')  # Margin boundary
plt.plot(xx, yy_up, 'k--')    # Margin boundary
plt.show()

# Predict and calculate accuracy
y_pred = svm.predict(X)
accuracy = accuracy_score(y, y_pred)
print("Accuracy: " + str(accuracy))
```
